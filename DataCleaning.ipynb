```python
import numpy as np
import pandas as pd
import math
```

```python
df = pd.read_csv('CP_HAM.csv')
df = df.drop('Comment',axis=1)
```

```python
df.head()
```

```python
def change_notation(row):
    if row['MotionParams.Side'] == 'Left':
        return 'L'
    if row['MotionParams.Side'] == 'Right':
        return 'R'
df['MotionParams.Side'] = df.apply(change_notation,axis=1)
df.head()
```

```python
uniquePID = df.iloc[:,0].dropna().unique()
print("Total unique patients: "+ str(len(uniquePID)))
```

```python
df2 = df.drop_duplicates('DeIdentifyNumber')
df2 = df2[['DeIdentifyNumber','SxCode']].dropna()

df2.shape
```

```python
def label_studyType(row):
    return df.loc[df['DeIdentifyNumber'] == row['DeIdentifyNumber'],'StudyType'].dropna().unique()
def number_of_study(row):
    return len(row['StudyType'])
def contains_post_op(row):
    return 'Post-op' in list(row['StudyType'])

df2['StudyType'] = df2.apply(label_studyType,axis=1)
df2['NumberOfStudy'] = df2.apply(number_of_study,axis=1)
df2['contains_post_op'] = df2.apply(contains_post_op,axis=1)
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2[ df2['NumberOfStudy'] == df2.NumberOfStudy.max()]
```

```python
df2 = df2.loc[df2['NumberOfStudy']>1 ]
df2 = df2.loc[df2['contains_post_op'] == True]
df2.shape
```

### List of unique patients with more than one study type which includes post-op study

```python
_uniquePIDlist= list(df2.DeIdentifyNumber)
```

### Dataframe having only surgical side data

```python
new_df = df.loc[df['Procedures.Side'] == df['MotionParams.Side']]
```

```python
print("Actual Dataframe shape: ",df.shape)
print("DataFrame with only surgical side data:",new_df.shape)
```

### Dataframe of patients with surgical side data and more than one study type 

```python
result_df = new_df[new_df['DeIdentifyNumber'].isin(_uniquePIDlist)]
```

```python
result_df.to_csv('CP_HAM_MODIFIED.csv',encoding='utf-8', index=False)
```
